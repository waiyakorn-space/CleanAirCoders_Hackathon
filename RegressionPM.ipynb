```python
import pandas as pd
from dataprep.eda import create_report
df = pd.read_csv("PMdata.csv")
```

```python
import numpy as np
df['winddir'] = df['winddir']* (np.pi / 180)
```

```python
df['Fire_BKK'] = df['Fire_brightness_BKK']*df['Fire_count_BKK']* df['winddir'] * df['windspeed']
df['Fire_CCO'] = df['Fire_brightness_CCO']*df['Fire_count_CCO']* df['winddir'] * df['windspeed']
df['Fire_NPT'] = df['Fire_brightness_NPT']*df['Fire_count_NPT']* df['winddir'] * df['windspeed']
df['Fire_NYK'] = df['Fire_brightness_NYK']*df['Fire_count_NYK']* df['winddir'] * df['windspeed']
df['Fire_NBI'] = df['Fire_brightness_NBI']*df['Fire_count_NBI']* df['winddir'] * df['windspeed']
df['Fire_PTE'] = df['Fire_brightness_PTE']*df['Fire_count_PTE']* df['winddir'] * df['windspeed']
df['Fire_SKN'] = df['Fire_brightness_SKN']*df['Fire_count_SKN']* df['winddir'] * df['windspeed']
df['Fire_SPK'] = df['Fire_brightness_SPK']*df['Fire_count_SPK']* df['winddir'] * df['windspeed']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_norm = df.iloc[:, 1:]
# Fit and transform the DataFrame
normalized_data = scaler.fit_transform(df_norm)

# Create a new DataFrame with the normalized data
normalized_df = pd.DataFrame(normalized_data, columns=df_norm.columns)
```

```python
normalized_df = normalized_df.drop(['visibility','Traffic00', 'Traffic01', 'Traffic02', 'Traffic03',
       'Traffic04', 'Traffic05', 'Traffic06', 'Traffic07', 'Traffic08',
       'Traffic09', 'Traffic10', 'Traffic11', 'Traffic12', 'Traffic13',
       'Traffic14', 'Traffic15', 'Traffic16', 'Traffic17', 'Traffic18',
       'Traffic19', 'Traffic20', 'Traffic21', 'Traffic22', 'Traffic23'], axis=1)
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
import numpy as np

# Importing LazyRegressor
from lazypredict.Supervised import LazyRegressor
from sklearn.utils import all_estimators
from sklearn.base import RegressorMixin
import xgboost
import lightgbm

removed_regressors = [
    "TheilSenRegressor",
    "ARDRegression",
    "CCA",
    "IsotonicRegression",
    "StackingRegressor",
    "MultiOutputRegressor",
    "MultiTaskElasticNet",
    "MultiTaskElasticNetCV",
    "MultiTaskLasso",
    "MultiTaskLassoCV",
    "PLSCanonical",
    "PLSRegression",
    "RadiusNeighborsRegressor",
    "RegressorChain",
    "VotingRegressor",
    'PoissonRegressor',
    'PassiveAggressiveRegressor',
    'RandomForestRegressor',
    'QuantileRegressor'
]

REGRESSORS = [
    est
    for est in all_estimators()
    if (issubclass(est[1], RegressorMixin) and (est[0] not in removed_regressors))
]

bool_series = pd.notnull(normalized_df["PM25"])

normalized_df = normalized_df[bool_series]
```

```python
X = normalized_df[['temp', 'humidity', 'precip', 'windspeed', 'winddir',
       'pres', 'o3', 'no2', 'so2', 'co', 'Fire_BKK', 'Fire_CCO', 'Fire_NPT', 'Fire_NYK', 'Fire_NBI',
       'Fire_PTE', 'Fire_SKN', 'Fire_SPK']]
```

```python
X.columns = ['อุณหภูมิ','ความชื้นสัมพัทธ์',"น้ำฝน","ความเร็วลม","ทิศทางลม","ความกดอากาศ","O3","NO2","SO2","CO","จุดไฟในกรุงเทพ","จุดไฟในฉะเชิงเทรา","จุดไฟในนครปฐม","จุดไฟในนครนายก","จุดไฟในนนทบุรี","จุดไฟในปทุมธานี","จุดไฟในสมุทรสาคร","จุดไฟในสมุทรปราการ"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, normalized_df['PM25'], test_size=0.33, random_state=42)
```

```python
# fitting data in LazyRegressor because
# here we are solving Regression use case.
reg = LazyRegressor(verbose=1, ignore_warnings=False, custom_metric=None, regressors=REGRESSORS)

# fitting data in LazyClassifier
models, predictions = reg.fit(X_train, X_test,
                              y_train, y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.inspection import permutation_importance
import numpy as np

# Create and train the HistGradientBoostingRegressor model
model = HistGradientBoostingRegressor()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Calculate feature importance using permutation importance
perm_importance = permutation_importance(model, X_test, y_test, n_repeats=30, random_state=8)
```

```python
# Get the feature importances and feature names
feature_importances = perm_importance.importances_mean
feature_names = X_test.columns  # Replace with your actual feature names

# Sort feature importances in descending order
sorted_indices = feature_importances.argsort()[::-1]
sorted_feature_importances = feature_importances[sorted_indices]
sorted_feature_names = [feature_names[i] for i in sorted_indices]

# Print sorted feature importances
print("Sorted Feature Importances:")
for feature_name, importance in zip(sorted_feature_names, sorted_feature_importances):
    print(f"Feature: {feature_name}, Importance: {importance}")
```

```python
import plotly.express as px
import pandas as pd

# Create a DataFrame with sorted feature names and importances
sorted_feature_df = pd.DataFrame({'ปัจจัย': sorted_feature_names, "ค่านัยสำคัญ": sorted_feature_importances})
sorted_feature_df["ค่านัยสำคัญ"] = sorted_feature_df["ค่านัยสำคัญ"].apply(lambda x: round(x, 4))

# Create an interactive bar plot using plotly.express
fig = px.bar(sorted_feature_df, x="ค่านัยสำคัญ", y='ปัจจัย', orientation='h',
             title='ความสัมพันธ์ของPM 2.5 กับปัจจัยต่างๆ', labels={"Feature": 'ค่านัยสำคัญ'})

# Customize the layout as needed
fig.update_layout(yaxis_categoryorder='total ascending', yaxis_title_text='ปัจจัย',
                  xaxis_title_text="ค่านัยสำคัญ")

# Show the plot
fig.show()
```
